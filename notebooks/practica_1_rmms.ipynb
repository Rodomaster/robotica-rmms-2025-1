### Definición de URDF.

El formato URDF (Unified Robot Description Format) es una especificación basada en XML que se utiliza para describir la estructura y las propiedades de un robot. A través de elementos definidos en XML, es posible especificar detalles clave del robot, como sus eslabones, juntas, geometría, materiales, propiedades inerciales y límites de movimiento. Este formato es ampliamente utilizado en simulación y visualización, especialmente en plataformas como ROS y RViz, permitiendo representar tanto el aspecto físico del robot como su comportamiento cinemático en entornos de simulación [[1]](#1).

- Aspecto visual: Define cómo se ve el robot, especificando la geometría, materiales y colores de cada uno de sus eslabones.
- Cinemática y dinámica: Establece cómo se mueven los eslabones del robot, detallando las juntas, los grados de libertad, las propiedades inerciales y los límites de movimiento.
- Modelo de colisión: Determina las formas y superficies que se utilizan para detectar colisiones, lo cual es esencial para simulaciones físicas y la interacción con otros objetos en un entorno virtual.
### Tipos de juntas que están definidas dentro del formato URDF.

Cada una de las partes rígidas que componen el robot se denomina eslabón o link. La forma en que se conectan dos eslabones se denomina articulación o joint [[1]](#1).

Dependiendo del tipo de movimiento que permita la conexión entre los eslabones, existen diferentes tipos de articulaciones en el formato URDF:

- revolute: permite el giro limitado sobre un eje, con límites definidos por valores superior e inferior.
- continuous: permite el giro sin límites sobre un eje, es decir, no tiene restricciones de rotación.
- prismatic: permite el deslizamiento a lo largo de un eje, con un rango limitado entre los valores superior e inferior.
- fixed: no permite ningún movimiento. No es una verdadera articulación, pero es útil para fijar un sistema de referencia a un eslabón, como para sensores o cámaras.
- floating: permite movimiento libre en los seis grados de libertad (traslación y rotación).
planar: permite el movimiento en un plano perpendicular a un eje específico.

Cuando dos eslabones están conectados por una articulación, el eslabón más cercano a la base o raíz del robot se denomina padre, mientras que el otro eslabón se llama hijo. Esta relación padre-hijo es fundamental para definir la estructura cinemática del robot en un entorno simulado o visualizado [[1]](#1).
### Definición de elementos "primitivos" que se pueden utilizar para definir un robot en ROS.

Cuando se solicita definir los tipos de elementos "primitivos" en ROS para describir un robot, se refiere a las formas geométricas básicas que se utilizan para definir las partes físicas del robot. Estos elementos son fundamentales para la visualización, simulación de colisiones y cálculos dinámicos [[2]](#2).

En URDF, los elementos geométricos "primitivos" disponibles incluyen:

- box: Define una caja o paralelepípedo rectangular mediante sus dimensiones de largo, ancho y alto.
- cylinder: Representa un cilindro, especificando el radio y la altura.
- sphere: Define una esfera a través de su radio.
- mesh: Permite usar formas más complejas, importadas de archivos externos (como STL o DAE), que no pueden describirse con las primitivas geométricas.


Estos elementos se emplean para modelar la geometría de los eslabones de manera sencilla y eficiente [[2]](#2).

### Que es un nodo y que es un topico en ROS 2
**Nodos** 
son programas ejecutables que realizan tareas específicas, como acceder a un motor, mapear el entorno o planificar rutas. Una de las principales ventajas de utilizar nodos es su diseño modular [[1]](#1). Cada nodo se compila de forma independiente y es gestionado por un nodo principal conocido como ROS Master. La comunicación entre nodos se realiza mediante mensajes y tópicos, lo que permite trabajar hacia un objetivo común. Gracias a su diseño modular y a la independencia de cada nodo, los proyectos se pueden abordar de manera mucho más sencilla  [[1]](#1).

Cada nodo está escrito utilizando la biblioteca cliente de ROS, que es un conjunto de herramientas de código que facilita el trabajo de los programadores en este entorno [[1]](#1). . Con esta biblioteca, es posible crear nodos, así como publicar y suscribirse a mensajes y tópicos, además de gestionar servicios y sus parámetros. Esta biblioteca es compatible con C++ y Python [[1]](#1).

**Topics**
Como se mencionó anteriormente, los nodos se comunican mediante el uso de Topics ("asuntos") a través de mensajes. La arquitectura de comunicaciones en ROS se basa en este sistema de publicación y suscripción, que permite una interacción eficiente entre nodos [[1]](#1).

Publicación (Publish): Un nodo envía (publica) un mensaje en un Topic específico. Este mensaje puede contener datos sobre el estado de un sensor, comandos de control o cualquier otra información relevante que deba ser compartida con otros nodos [[1]](#1).

Suscripción (Subscribe): Varios nodos pueden suscribirse a ese mismo Topic para recibir los mensajes publicados. Al hacerlo, obtienen acceso a la información en tiempo real y pueden actuar en consecuencia [[1]](#1).

Esta arquitectura de comunicación permite que los nodos funcionen de manera descentralizada y facilita la escalabilidad del sistema. Al utilizar Topics, se evita la necesidad de que los nodos conozcan la existencia de otros nodos directamente, lo que reduce el acoplamiento entre ellos. Esto permite que el sistema sea más flexible y modular, ya que los nodos pueden ser añadidos o eliminados sin afectar la funcionalidad general. Además, el uso de Topics es fundamental para manejar la naturaleza asincrónica de las comunicaciones en sistemas robóticos, donde los eventos pueden ocurrir en diferentes momentos [[1]](#1).

### Que es RVIZ.

RViz es una herramienta de software para visualización y depuración de robots. Le permite mostrar e interactuar con varios tipos de datos, como imágenes, nubes de puntos, mapas, trayectorias, etc. También puede usar RViz para interactuar con ROS y suscribirse y publicar temas, servicios y acciones de ROS [[3]](#3).

## Desarrollo

En la práctica realizada, primero se tuvo que establecer el espacio de trabajo, lo cual se hizo desde la terminal. Después de crear el espacio de trabajo con las carpetas necesarias para empezar a trabajar, abrimos Visual Studio Code para redactar el documento URDF. Iniciamos definiendo el nombre del robot y creando una junta, la cual solo sirve para establecer que esta está fija en el suelo y en el origen. Procedimos a establecer nuestro primer eslabón, al que llamamos "basic link", definiendo su origen y su geometría como una caja con dimensiones que nosotros especificamos. También definimos su color, que es gris, y sus relaciones de colisión. Luego describimos la inercia del eslabón y su matriz de inercia en su centro de masa. Estos parámetros son similares en los demás eslabones, llamados "link 1" y "link 2", cambiando solo algunos aspectos como su origen, tamaño, material y color.

Una vez definido esto, continuamos con la definición de las juntas, donde primero establecemos su nombre y el tipo de junta. La junta para el eslabón 1 es una junta rotacional; se especificó el padre y el hijo de la junta, que indica de dónde a dónde estará situada. Definimos el eje de movimiento, los límites de movimiento y su posición, que debe estar desplazada del origen considerando la ubicación del siguiente eslabón respecto a la junta anterior. Este proceso se repitió con la otra junta, teniendo en cuenta que es de tipo prismática y que tendrá otro eje por el cual se desplazará.

Con esto completado, se establecieron los otros documentos necesarios para ejecutar RViz. Se modificó el documento CMakeLists para que, con la instrucción de install, se ejecutaran los paquetes de launch, URDF y RViz. Creamos en la carpeta de launch un documento que permite automatizar la ejecución de múltiples nodos y configuraciones necesarias para el funcionamiento de un sistema robótico. Finalmente, realizamos el documento package, que es crucial para definir los componentes de un paquete y gestionar su correcta instalación y ejecución dentro de un entorno ROS 2.

Finalmente, compilamos con el comando colcon build y ejecutamos la instrucción source install/setup.bash para poder interactuar con los paquetes y herramientas de ROS 2 en el espacio de trabajo. Por último, ejecutamos el programa de launch, lo que nos permitió abrir RViz con el comportamiento y las características que configuramos en el URDF.
